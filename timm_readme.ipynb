### &ensp;Load model without the final layer (fc)
#### &emsp;create_model의 argument를 조절
&emsp;&emsp;`drop_rate` : set the dropout rate for training (Default: `0`)<br>
&emsp;&emsp;`num_classes` : the number of output neurons corresponding to classes

```python
import torch
import timm
m = timm.create_model('resnet50', pretrained=True, num_classes=0)
m.global_pool
```

&ensp;`global_pool` : determine the type of global pooling to be used before the final classification layers
 - avg : Average pooling
 - max : Max pooling
 - avgmax: the sum of average and max pooling, re-scaled by 0.5
 - catavgmax: a concatenation of the outputs of average and max pooling along feature dimension. Note that this will double the feature dimension.
 - ‘’ : No pooling is used, the pooling layer is replaced by an Identity operation

```python
pool_types = ['avg', 'max', 'avgmax', 'catavgmax', '']
```

```python
for pool in pool_types:
    m = timm.create_model('resnet50', pretrained=True, num_classes=0, global_pool=pool)
    o = m(torch.randn(2, 3, 224, 224))
    print(f'Unpooled shape: {o.shape}')            
    m.global_pool
```

### Final Feature Map (final convolution layer)
#### &emsp; bypassing the global pooling and classification layer.

```python
import torch
import timm

model = timm.create_model('resnet50', pretrained=True)
```

```python
import matplotlib.pyplot as plt
def visualize_feature_output(t):
    plt.imshow(t[0].transpose(0,2).sum(-1).detach().numpy())
    plt.show()
```

```python
import glob
import resize
import numpy as np

image_path_list = glob.glob("/media/nd-ygr/Data/Image/google/dataset/v6/images/Dog/test/*")
image = resize.vips_resize(image_path_list[0], (1080,1080), False)
image = torch.as_tensor(np.array(image, dtype=np.float32)).transpose(2,0)[None]
```

```python
model = timm.create_model('resnet50', pretrained=True)
feature_output= model.forward_features(image)
visualize_feature_output(feature_output)
```

### a model as a feature backbone
#### &emsp; Select specific feature levels or limit the stride
 &emsp; - `out_indices` selects which indices to output<br>
 &emsp; - `output_stride` limits the feature output stride of the network (also works in classification mode BTW)

```python
import timm

model = timm.create_model('resnet50d', pretrained = True, features_only = True)
```

```python
print(model.feature_info.module_name())
print(model.feature_info.channels())
print(model.feature_info.reduction())
```

```python
import glob
import resize
import numpy as np

image_path_list = glob.glob("/media/nd-ygr/Data/Image/google/dataset/v6/images/Dog/test/*")
image = resize.vips_resize(image_path_list[0], (1080,1080), False)
image = torch.as_tensor(np.array(image, dtype=np.float32)).transpose(2,0)[None]
```

```python
out = model(image)
for i, o in enumerate(out):
    if divmod(i+1,2)[1] == 0:
        continue
    print(f"Feature Map {i+1}")
    visualize_feature_output(o)
```
